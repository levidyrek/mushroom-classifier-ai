```python
import pandas as pd
mushroom_data = pd.read_csv('data/mushroom_data.csv', sep=';')
mushroom_data
```

```python
cols_to_norm = ['cap-diameter', 'stem-height']
mushroom_data[cols_to_norm] = mushroom_data[cols_to_norm].apply(lambda x: (x - x.min()) / (x.max() - x.min()))
```

```python
import tensorflow as tf

cap_diameter = tf.feature_column.numeric_column('cap-diameter')
cap_shape = tf.feature_column.categorical_column_with_vocabulary_list('cap-shape', ['b', 'c', 'x', 'f', 's', 'p' 'o'])
cap_surface = tf.feature_column.categorical_column_with_vocabulary_list('cap-surface', ['i', 'g', 'y', 's', 'h', 'l', 'k', 't', 'w', 'e'])
cap_color = tf.feature_column.categorical_column_with_vocabulary_list('cap-color', ['n', 'b', 'g', 'r', 'p', 'u', 'e', 'w', 'y', 'l', 'o', 'k'])
does_bruise_or_bleed = tf.feature_column.categorical_column_with_vocabulary_list('does-bruise-or-bleed', ['t', 'f'])

# TODO

stem_height = tf.feature_column.numeric_column('stem-height')
```
